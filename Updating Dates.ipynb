```python
import pandas as pd
import datetime
```

```python
mega_data = pd.read_csv("final_mega_df.csv")
mega_data
```

```python
temp = mega_data['earliest_cr_line']
temp
```

```python
#earliest_cr_line is a string, want to change to datetime format
#first, a dictionary comparing the string of month names to their numerical value

months = {
        'Jan' : 1,
        'Feb' : 2,
        'Mar' : 3,
        'Apr' : 4,
        'May' : 5,
        'Jun' : 6,
        'Jul' : 7,
        'Aug' : 8,
        'Sep' : 9, 
        'Oct' : 10,
        'Nov' : 11,
        'Dec' : 12
}

#save in separate field, and split to two columns
temp = mega_data['earliest_cr_line']
temp.str.split('-')
new = mega_data['earliest_cr_line'].str.split("-", n = 1, expand = True) 

#change to numerical value and save as Integers
new[0]= new[0].map(months) 
new[1]= new[1].astype(int)
```

```python
#change Year value to full numerical year by adding 2000
new[1] = 2000 + new[1]
dates = new
dates
```

```python
for i, row in dates.iterrows():
    if dates[1][i] > 2019:
        dates[1][i] = dates[1][i] - 100
        
    
```

```python
dates
```

```python
#rename columns
dates.rename(columns = {0: 'Month', 1: 'Year'}, inplace = True)

#convert to datetime format
dates['Full Date'] = pd.to_datetime(dates.assign(Day=1).loc[:, ['Year','Month','Day']])

#add column that just shows Year/Month values, as Day value doesn't matter
dates['earliest_cr_line']= pd.to_datetime(dates['Full Date']).dt.to_period('M')
dates
```

```python
#merge issue_dates into mega_data 

mega_data['credit_start_date']= dates['earliest_cr_line']

mega_data.head()
```

```python
credit_data = mega_data[['ID', 'loan_amnt', 'funded_amnt','int_rate','installment',
                       'annual_inc','fico_range_low','fico_range_high','dti',
                       'credit_start_date']]
credit_data
```
